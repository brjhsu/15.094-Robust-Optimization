## 2.1 & 2.2 Nominal Problem

I set up my problem as the following LP

\begin{equation}
\begin{aligned}
\min_{r,e,u} \quad & \sum_{t} \sum_{s} \left[ \sum_{i} F_{i}r_{i,s}+V_{i}e_{i,s,t} + \sum_{k} C_{i}u_{i,s,k,t} \right] \\
\textrm{s.t.} \quad & r_{i,s} + e_{i,s,t}+\sum_{k}u_{i,k,s,t}-\sum_{k}u_{i,s,k,t} \geq D_{i,s,t}  \quad \forall i,s,t\\
  & h_{s,1}=\sum_{i} e_{i,s,1} \quad \forall s \\
  & h_{s,2}=\frac{1}{2} \left( \sum_{i} e_{i,s,1} + \sum_{i}e_{i,s,2} \right) \quad \forall s \\
  & h_{s,t}=\frac{1}{3} \left( \sum_{i} e_{i,s,t-2} + \sum_{i}e_{i,s,t-1} +\sum_{i} e_{i,s,t} \right) \quad \forall s,t\geq3 \\
  & h_{s,t} \leq 1 \quad \forall s,t \\
  & r,e,u \geq 0 \\
\end{aligned}
\end{equation}

This is set up to minimize costs. We make sure that all demands are met with either the base capacity, the extended capacity, or the net transfered resources (transfered in - transfered out).

```julia
using JuMP, JuMPeR, Gurobi, Random, Distributions, LinearAlgebra, Plots, StatsPlots
```

```julia
T = 25
S = 5
I = 6
Random.seed!(MersenneTwister(314)) # Do not change the seed. 
function generate_data(I::Int64, S::Int64, T::Int64)
    mvdist = Multinomial(I, S) # Multinomial w.r.t. server allocation
    raws = Array{Any}(zeros(S)) # Generating tridiagonal multivariate normals for the servers
    for s = 1:S 
        randos = 1/5*randperm(S)
        mat = diagm(0 => 1*ones(S), 1 => randos, -1 => randos)
        symmat = mat * mat'
        raws[s] = MvNormal(symmat)
    end
    D = zeros(I, S, T); # Generating demand data
    for t = 1:T
        D[:,:, t] += 2*rand(mvdist,I)'
        for i = 1:I
            for s = 1:S
                D[i, s, t] += abs(rand(raws[s])[1])
            end
        end
    end
    return D
end
D = generate_data(I,S,T);
```

## 2.1 & 2.2 Nominal Problem

I set up my problem as the following LP

\begin{equation}
\begin{aligned}
\min_{r,e,u} \quad & \sum_{t} \sum_{s} \left[ \sum_{i} F_{i}r_{i,s}+V_{i}e_{i,s,t} + \sum_{k} C_{i}u_{i,s,k,t} \right] \\
\textrm{s.t.} \quad & r_{i,s} + e_{i,s,t}+\sum_{k}u_{i,k,s,t}-\sum_{k}u_{i,s,k,t} \geq D_{i,s,t}  \quad \forall i,s,t\\
  & h_{s,1}=\sum_{i} e_{i,s,1} \quad \forall s \\
  & h_{s,2}=\frac{1}{2} \left( \sum_{i} e_{i,s,1} + \sum_{i}e_{i,s,2} \right) \quad \forall s \\
  & h_{s,t}=\frac{1}{3} \left( \sum_{i} e_{i,s,t-2} + \sum_{i}e_{i,s,t-1} +\sum_{i} e_{i,s,t} \right) \quad \forall s,t\geq3 \\
  & h_{s,t} \leq 1 \quad \forall s,t \\
  & r,e,u \geq 0 \\
\end{aligned}
\end{equation}

This is set up to minimize costs. We make sure that all demands are met with either the base capacity, the extended capacity, or the net transfered resources (transfered in - transfered out). 

```julia
# OPTIMIZATION MODEL HERE.
# m = RobustModel(solver = GurobiSolver())
# nominal optimization model 
# D is (I resources) x (S servers) x (T timesteps)
m = Model(solver = GurobiSolver(OutputFlag=0))
@variable(m, r[1:I,1:S]>=0) # fixed resource capacity
@variable(m, e[1:I,1:S,1:T]>=0) # expansion
@variable(m, u[1:I,1:S,1:S,1:T]>=0) # amount to transfer from one source to another 
@variable(m, h[1:S,1:T] >= 0)
# u_{i,s,k,t} is the amount of resource "i" going from server "s" to server "k"

@constraint(m, [i=1:I, s=1:S, t=1:T], r[i,s]+e[i,s,t]+sum(u[i,s,:,t])-sum(u[i,:,s,t]) >= D[i,s,t])
@constraint(m, [s=1:S], h[s,1]==sum(e[:,s,1])) # t=1 heat capacity constraint 
@constraint(m, [s=1:S], h[s,2]==(1/2)*sum(e[:,s,1:2])) # t=2 heat capacity constraint 
@constraint(m, [s=1:S,t=3:T], h[s,t]==(1/3)*sum(e[:,s,t]+e[:,s,t-1]+e[:,s,t-2])) # t=3 heat capacity constraint 
@constraint(m, [s=1:S,t=1:T], h[s,t]<=1)

@objective(m, Min, sum(sum(sum(F[i]*r[i,s] + V[i]*e[i,s,t] + sum(C[i]*u[i,s,k,t] for k in 1:S) for i in 1:I) for s in 1:S) for t in 1:T));
```

```julia
solve(m)
println("Objective value: ", getobjectivevalue(m))
```

## 2.4 Robust Solution with budget uncertainty

Here we have $-1 \leq d_{i,s,t} \leq 1$ as an uncertainty parameter for each demand $D_{i,s,t}$ out of $L=I\times S \times T = $ 750 variables. The uncertainty is decoupled ($P=I$) and we need this to be satisfied with probability $1-\epsilon=0.95$. Under a budget uncertainty, this means we'll $\rho = 1$ and solve for $\Gamma$

```julia
Γ=sqrt(2*log(1/0.05)*I*S*T);
print("Γ value: ", Γ)
```

```julia
# OPTIMIZATION MODEL HERE.
m = RobustModel(solver = GurobiSolver(OutputFlag=0))
# nominal optimization model 
# D is (I resources) x (S servers) x (T timesteps)

@uncertain(m,d[1:I,1:S,1:T])
@uncertain(m,z[1:I,1:S,1:T])
@constraint(m,d.==D+z) # the uncertain demand is D+z where z can range [-1,1]
@constraint(m, norm(z,   1) <= Γ)  # ‖z‖₁ ≤ Γ
@constraint(m, norm(z, Inf) <= 1)  # ‖z‖∞ ≤ 1


@variable(m, r[1:I,1:S]>=0) # fixed resource capacity
@variable(m, e[1:I,1:S,1:T]>=0) # expansion
@variable(m, u[1:I,1:S,1:S,1:T]>=0) # amount to transfer from one source to another 
@variable(m, h[1:S,1:T] >= 0)
# u_{i,s,k,t} is the amount of resource "i" going from server "s" to server "k"

@constraint(m, [i=1:I, s=1:S, t=1:T], r[i,s]+e[i,s,t]+sum(u[i,s,:,t])-sum(u[i,:,s,t]) >= d[i,s,t]) # change from big D to d
@constraint(m, [s=1:S], h[s,1]==sum(e[:,s,1])) # t=1 heat capacity constraint 
@constraint(m, [s=1:S], h[s,2]==(1/2)*sum(e[:,s,1:2])) # t=2 heat capacity constraint 
@constraint(m, [s=1:S,t=3:T], h[s,t]==(1/3)*sum(e[:,s,t]+e[:,s,t-1]+e[:,s,t-2])) # t=3 heat capacity constraint 
@constraint(m, [s=1:S,t=1:T], h[s,t]<=1)

@objective(m, Min, sum(sum(sum(F[i]*r[i,s] + V[i]*e[i,s,t] + sum(C[i]*u[i,s,k,t] for k in 1:S) for i in 1:I) for s in 1:S) for t in 1:T));
```
